Ejemplo 1: Ventana con boton que cambia el texto de una etiqueta 

```python
import tkinter as tk  # Se importa la biblioteca tkinter con el alias 'tk' para crear interfaces gráficas.

def cambiar_texto():
    etiqueta.config(text="Texto cambiado")  # Esta función cambia el texto mostrado en la etiqueta.

ventana = tk.Tk()  # Se crea la ventana principal de la aplicación.
ventana.title("Ejemplo 1")  # Se establece el título de la ventana.

etiqueta = tk.Label(ventana, text="Texto original")  # Se crea una etiqueta con el texto inicial "Texto original".
etiqueta.pack()  # Se coloca la etiqueta en la ventana usando el gestor de geometría 'pack'.

boton = tk.Button(ventana, text="Cambiar texto", command=cambiar_texto)  # Se crea un botón con el texto "Cambiar texto".
                                                                         # Al hacer clic, ejecuta la función 'cambiar_texto'.
boton.pack()  # Se coloca el botón en la ventana.

ventana.mainloop()  # Se inicia el bucle principal de la aplicación.
                    # Esto hace que la ventana se mantenga abierta y responda a eventos (como clics).
```

Ejemplo 2: Entrada de texto y botón para mostrar mensaje

```python
import tkinter as tk  # Se importa la biblioteca tkinter con el alias 'tk' para crear interfaces gráficas.

# Función que se ejecuta al presionar el botón "Mostrar".
# Toma el texto que el usuario escribió en la entrada y lo muestra en la etiqueta.
def mostrar():
    etiqueta.config(text=f"Ingresaste: {entrada.get()}")  # Cambia el texto de la etiqueta usando lo ingresado en la caja de texto.

ventana = tk.Tk()  # Se crea la ventana principal de la aplicación.
ventana.title("Ejemplo 2")  # Se establece el título de la ventana.

entrada = tk.Entry(ventana)  # Se crea una caja de texto para que el usuario escriba.
entrada.pack()  # Se coloca la caja de texto en la ventana.

boton = tk.Button(ventana, text="Mostrar", command=mostrar)  # Se crea un botón con el texto "Mostrar".
                                                              # Al hacer clic, ejecuta la función 'mostrar'.
boton.pack()  # Se coloca el botón debajo de la caja de texto.

etiqueta = tk.Label(ventana, text="Esperando entrada")  # Se crea una etiqueta con el texto inicial "Esperando entrada".
etiqueta.pack()  # Se coloca la etiqueta en la ventana, debajo del botón.

ventana.mainloop()  # Se inicia el bucle principal de la aplicación.
                    # Este mantiene la ventana abierta y lista para interactuar con el usuario.
```

```python
import tkinter as tk  # Se importa la biblioteca tkinter con el alias 'tk'.

# Función para mostrar el texto ingresado.
def mostrar():
    texto = entrada.get().strip()  # Obtiene el texto ingresado, eliminando espacios.
    if texto == "":
        etiqueta.config(text=" Por favor, ingresa un texto.")  # Mensaje si está vacío.
    else:
        etiqueta.config(text=f" Ingresaste: {texto}")  # Muestra el texto en la etiqueta.

# Función para limpiar la entrada y la etiqueta.
def limpiar():
    entrada.delete(0, tk.END)  # Borra todo el contenido de la entrada.
    etiqueta.config(text="Esperando entrada...")  # Restaura el mensaje original.

# Crear la ventana principal.
ventana = tk.Tk()  # Se crea una instancia de la ventana principal de la interfaz.
ventana.title("Ejemplo Modificado")  # Se establece el título de la ventana.
ventana.geometry("350x200")  # Se define el tamaño de la ventana: 350 píxeles de ancho por 200 de alto.

# Caja de texto para entrada.
entrada = tk.Entry(ventana, font=("Arial", 12), width=30)  # Se crea un campo de entrada de texto con fuente Arial tamaño 12 y ancho de 30 caracteres.
entrada.pack(pady=10)  # Se coloca la entrada en la ventana con un espacio vertical (padding) de 10 píxeles.

# Botón para mostrar el texto ingresado.
boton_mostrar = tk.Button(
    ventana,                  # Se especifica que el botón estará dentro de la ventana principal.
    text="Mostrar",           # Texto que se mostrará en el botón.
    command=mostrar,          # Función que se ejecutará cuando se presione el botón.bg="lightgreen",             # Color de fondo del botón (verde claro).
    font=("Arial", 11)        # Fuente y tamaño del texto del botón.
)
boton_mostrar.pack(pady=5)  # Se coloca el botón con un espacio vertical de 5 píxeles.

# Botón para limpiar la entrada.
boton_limpiar = tk.Button(
    ventana,                  # Se coloca también en la ventana principal.
    text="Limpiar",           # Texto que aparece en el botón.
    command=limpiar,          # Función que se ejecutará al hacer clic.
    bg="lightcoral",             # Color de fondo del botón (rojo claro).
    font=("Arial", 11)        # Fuente y tamaño del texto del botón.
)
boton_limpiar.pack(pady=5)  # Se coloca con un pequeño espacio vertical.

# Etiqueta que muestra mensajes o el texto ingresado.
etiqueta = tk.Label(
    ventana,                          # La etiqueta se coloca dentro de la ventana principal.
    text="Esperando entrada...",      # Texto inicial de la etiqueta.
    font=("Arial", 12),               # Fuente del texto de la etiqueta.
    fg="blue"                         # Color del texto de la etiqueta.
)
etiqueta.pack(pady=10)  # Se coloca la etiqueta con espacio vertical.

# Ejecutar la aplicación.
ventana.mainloop()  # Inicia el bucle principal de la aplicación.
                    # Este bucle mantiene abierta la ventana y a la espera de eventos (como clics o escritura).
```

Ejemplo 3 : Lista desplegable para seleccionar una opción

```python
import tkinter as tk  # Se importa la biblioteca tkinter con el alias 'tk'.

# Función que se ejecuta cuando el usuario selecciona una opción del menú desplegable.
def seleccionar(opcion):
    etiqueta.config(text=f"Seleccionaste: {opcion}")  # Cambia el texto de la etiqueta para mostrar la opción seleccionada.

ventana = tk.Tk()  # Se crea la ventana principal de la aplicación.
ventana.title("Ejemplo 3")  # Se establece el título de la ventana.

# Lista de opciones para el menú desplegable.
opciones = ["Rojo", "Verde", "Azul"]

# Creamos una variable asociada a la selección del menú.
variable = tk.StringVar(ventana)
variable.set(opciones[0])  # Establecemos la opción inicial (la primera opción de la lista).

# Crea


r el menú desplegable (OptionMenu) con las opciones disponibles y la variable que almacena la opción seleccionada.
menu = tk.OptionMenu(ventana, variable, *opciones, command=seleccionar)
menu.pack()  # Se coloca el menú desplegable en la ventana.

# Etiqueta que muestra un mensaje de instrucción o la opción seleccionada por el usuario.
etiqueta = tk.Label(ventana, text="Elige una opción")
etiqueta.pack()  # Se coloca la etiqueta en la ventana.

# Iniciar la aplicación y mantener la ventana abierta para interacción del usuario.
ventana.mainloop()  # Esto mantiene la ventana abierta y esperando eventos (como seleccionar una opción).
```

Ejemplo 3 modificado

```python
import tkinter as tk  # Se importa la biblioteca tkinter con el alias 'tk' para crear interfaces gráficas.

# Función que se ejecuta cuando el usuario selecciona una opción del menú desplegable.
def seleccionar(opcion):
    # Cambiar el color de fondo de la ventana según la opción seleccionada.
    if opcion == "Rojo":
        ventana.config(bg="red")  # Cambia el color de fondo de la ventana a rojo.
    elif opcion == "Verde":
        ventana.config(bg="green")  # Cambia el color de fondo de la ventana a verde.
    elif opcion == "Azul":
        ventana.config(bg="blue")  # Cambia el color de fondo de la ventana a azul.
    
    # Cambiar el texto de la etiqueta para mostrar la opción seleccionada.
    etiqueta.config(text=f"Seleccionaste: {opcion}")  # Actualiza el texto de la etiqueta.

# Crear la ventana principal.
ventana = tk.Tk()  # Se crea la ventana principal de la aplicación.
ventana.title("Ejemplo 3 Modificado")  # Se establece el título de la ventana.
ventana.geometry("300x200")  # Establecer un tamaño adecuado para la ventana.

# Lista de opciones para el menú desplegable.
opciones = ["Rojo", "Verde", "Azul"]

# Crear una variable asociada a la selección del menú.
variable = tk.StringVar(ventana)
variable.set(opciones[0])  # Establecemos la opción inicial.

# Crear el menú desplegable con las opciones disponibles y la variable que almacena la opción seleccionada.
menu = tk.OptionMenu(ventana, variable, *opciones, command=lambda x: seleccionar(x))  # Se pasa el valor correctamente a la función
menu.config(font=("Arial", 12), bg="lightblue")  # Configuramos el estilo del menú desplegable.
menu.pack(pady=15)  # Se coloca el menú desplegable en la ventana con un espaciado vertical de 15 píxeles.

# Etiqueta que muestra un mensaje de instrucción o la opción seleccionada por el usuario.
etiqueta = tk.Label(ventana, text="Elige una opción", font=("Arial", 12), fg="darkblue")
etiqueta.pack(pady=10)  # Se coloca la etiqueta en la ventana con un espaciado vertical de 10 píxeles.

# Etiqueta que muestra el mensaje de la opción seleccionada.
etiqueta_seleccion = tk.Label(ventana, text="Selecciona una opción del menú.", font=("Arial", 12), fg="green")
etiqueta_seleccion.pack(pady=10)  # Se coloca la etiqueta para mostrar la selección.

# Iniciar la aplicación y mantener la ventana abierta para interacción del usuario.
ventana.mainloop()  # Esto mantiene la ventana abierta y esperando eventos.
```
